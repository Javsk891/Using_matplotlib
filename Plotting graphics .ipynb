```python
# Times series of rainfall fron de different subdivisions in india (1901-2015)
rain1 = pd.read_csv('C:\\Users\\USUARIO\\Desktop\\JAVO\\Python for Data Analisis\\scriptsLecture\\scriptsLecture\\section6\\RainIndia.csv')
rain1.head(7)
```
